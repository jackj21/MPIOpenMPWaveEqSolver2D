```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
#512 MPI Tasks - 1st exp
df = pd.read_csv('wave_time_all_q5.csv', header=0, sep=' ')
df2 = df.groupby(df['n']).mean()
df2
```

```python
df2.index = pow(df2.index, 2)
df2
```

```python
plt.loglog(df2.index, df2['time'])
plt.title("OpenMP/MPI Combo Wave Time Simulator - 512 MPI Tasks")
plt.xlabel("N")
plt.ylabel("Avg. Run Time (s)")
```

```python
#256 MPI Tasks - 2nd experiment
df_b = pd.read_csv('wave_time_all_q5_b.csv', header=0, sep=' ')
df2b = df_b.groupby(df_b['n']).mean()
df2b
```

```python
df2b.index = pow(df2b.index, 2)
df2b
```

```python
df4 = pd.read_csv('wave_time_all.csv', header=0, sep=' ')
df4 = df.groupby(df['n']).mean()
df4
```

```python
df4.index = pow(df4.index, 2)
df4
```

```python
plt.loglog(df2.index, df2['time'], label='512 MPI Tasks')
plt.loglog(df2b.index, df2b['time'], label = '256 MPI Tasks')
plt.loglog(df4.index, df4['time'], label = 'Project 4')
plt.legend()
plt.title("OpenMP/MPI Combo Wave Time Simulator - 256 MPI Tasks")
plt.xlabel("N")
plt.ylabel("Avg. Run Time (s)")
```
